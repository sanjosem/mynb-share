```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
mpl.rcParams['font.size']=22
```

Comparaison between several experiment results:
* Gyuzel - clean, serrated and background recorded at 1.0 m
* Prateek - clean at 8 deg, measurements recorded at 1.21m
* Prateek - clean at 5 deg, measurements recorded at 1.34m

Everything is rescaled at 1 m.

```python
# Exp Gyuzel XX deg, 1m
data_Gyuzel = np.loadtxt('Data_u16.txt',skiprows=1,delimiter=',')
header=['Frequency','Background','Clean','Serrated']
```

```python
# Expe clean 8 deg 1.21m
data_Prateek_8deg = np.loadtxt('SPL_f_PSD.dat',skiprows=1)
```

```python
# Expe clean at 1.36 m 
data_Prateek_5deg = np.loadtxt('spectra_farfield_ss.dat',skiprows=0)
```

```python
plt.figure(figsize=(12,8))
for icol in range(1,data_Gyuzel.shape[1]):
    plt.plot(data_Gyuzel[:,0],10*np.log10(data_Gyuzel[:,icol]/4.0e-10),label=header[icol])
plt.plot(data_Prateek_5deg[:,1],10*np.log10(data_Prateek_5deg[:,0]/4.0e-10)+20.*np.log10(1.21/1.0),
         label=r'Prateek - 5$^\circ$')
plt.plot(data_Prateek_8deg[:,0],data_Prateek_8deg[:,1]+20.*np.log10(1.36/1.0),label=r'Prateek - 8$^\circ$')
plt.xscale('log')
plt.xlim(100,20000)
plt.ylim(-20,50)
plt.grid()
plt.ylabel(r'$S_{pp}$ [dB @ 1Hz, 1m, ref 2.0$\mu$Pa]')
plt.xlabel(r'Frequency [Hz]');
plt.legend();
```
